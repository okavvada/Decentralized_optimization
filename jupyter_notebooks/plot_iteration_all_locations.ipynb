```python
from __future__ import division
import csv
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import pylab as pl
get_ipython().magic(u'matplotlib inline')
```

```python
data_financial = pd.read_csv('../iteration_results_financial.csv', header=None, skiprows = 1)
data_sunset = pd.read_csv('../iteration_results_sunset.csv', header=None, skiprows = 1)
data_twin = pd.read_csv('../iteration_results_twin.csv', header=None, skiprows = 1)
data_richmond = pd.read_csv('../iteration_results_richmond.csv', header=None, skiprows = 1)
```

```python
cols = ['index', 'index_2','res_people', 'commercial_people', 'metric', 'in_build_pump', 'treat', 'treat_emb', 'piping']
data_financial.columns = cols
data_sunset.columns = cols
data_twin.columns = cols
data_richmond.columns = cols
data_financial['total_people'] = data_financial['res_people'] + data_financial['commercial_people'] 
data_sunset['total_people'] = data_sunset['res_people'] + data_sunset['commercial_people'] 
data_twin['total_people'] = data_twin['res_people'] + data_twin['commercial_people'] 
data_richmond['total_people'] = data_richmond['res_people'] + data_richmond['commercial_people'] 
data_financial['people_sum'] = data_financial.total_people.cumsum()
data_sunset['people_sum'] = data_sunset.total_people.cumsum()
data_twin['people_sum'] = data_twin.total_people.cumsum()
data_richmond['people_sum'] = data_richmond.total_people.cumsum()
```

```python
data_financial = data_financial[1:]
plt.scatter(data_financial['people_sum'], data_financial['metric'], color = 'r', alpha=0.7, s=50)
plt.xlabel('Scale (# of people)',fontsize =16)
plt.ylabel('Energy ($kWh/m^3$)',fontsize =16)
plt.ylim(6,16)
plt.xlim(-500,10000)
plt.xticks(fontsize =16)
plt.yticks(fontsize =16)
plt.grid(color='k', linestyle='--', linewidth=0.2, axis='both', alpha = 0.7)
```

```python
data_plot = data_financial[['in_build_pump', 'treat', 'treat_emb', 'piping']]
data_plot.columns = [['pumping', 'treatment operation', 'treatment capital', 'piping']]
data_plot.plot(kind='bar', stacked = True, colormap='PRGn', legend=False)
plt.xticks(np.arange(len(data_plot)), map(int, data_financial['people_sum']))
plt.ylabel('Energy ($kWh/m^3$)',fontsize =14)
plt.yticks(fontsize =14)
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off
```
